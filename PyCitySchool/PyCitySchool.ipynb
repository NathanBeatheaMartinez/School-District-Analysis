```python
# -----------------------------------------------------------------------------------------
# DISTRICT SUMMARY
# -----------------------------------------------------------------------------------------
```

```python

# create a list of all the value in 'school_name' column
schools_list = df.iloc[:, [4]]

# create a list of the unique value in "school_name" column
unique_schools_list = df.school_name.unique()

#Create a variable to hold the total count of schools in data frame
schools_count = len(unique_schools_list)

# create a variable to hold average_math_score use .mean() function to find average
avg_math_score = round(df.math_score.mean(), 2)

# create a variable to hold average_reading_score use .mean() function to find average
avg_reading_score = round(df.reading_score.mean(), 2)

# create a list that isolates whether the student passed or failed math (True or False)
passed_math_list = df.iloc[:, [6]] >= 70

# create a list that isolates whether the student passed or failed reading (True or False)
passed_reading_list = df.iloc[:, [5]] >= 70

# create a list that isolates whether the student passed math and reading (True or False)
dfboth = df.loc[(df['math_score'] >= 70) & (df['reading_score'] >= 70)]

passed_both_list = dfboth.iloc[:, [5]]

# use count function over 'passed_math_list' to find count of students who passed math
passed_math_count = passed_math_list[passed_math_list.math_score == True].count()

# create passed_math_pct variable to hold percentage of students who passed math
passed_math_pct = round((passed_math_count/students_count) * 100, 2)

# use count function over 'passed_reading_list' to find count of students who passed reading
passed_reading_count = passed_reading_list[passed_reading_list.reading_score == True].count()

# create passed_reading_pct variable to hold percentage of students who passed reading
passed_reading_pct = round((passed_reading_count/students_count) * 100, 2)

# create a variable for pass_both_count which will hold the total number of students who passed both math and reading
passed_both_count = len(passed_both_list)

# create passed_both_pct variable to hold percentage of students who passed math and reading
passed_both_pct = round((passed_both_count/students_count)*100, 2)

# use count function to count the total value in 'student name' column (total number of students)
students_count = df.student_name.count()

# create a list 'unique_budgets_list' to hold each unique budget
unique_budgets_list = df.budget.unique()

# use sum function over the 'budget column' to get the total budget for the district
total_budget = unique_budgets_list.sum()

# add value to a seperate "District Summary" dataframe
district_sum_df = pd.DataFrame(
                            {"Total Schools" : [schools_count],
                             "Total Students" : [students_count],
                             "Total Budget" : [total_budget],
                             "Average Math Score" : [avg_math_score],
                             "Average Reading Score" : [avg_reading_score],
                             "% Passing Math" : [passed_math_pct],
                             "% Passing Reading" : [passed_reading_pct],
                             "% Overall Passing" : [passed_both_pct]})

# Formatting
district_sum_df["Total Students"] = district_sum_df["Total Students"].map("{:,}".format)
district_sum_df["Total Budget"] = district_sum_df["Total Budget"].map("${:,.2f}".format)

# Display
district_sum_df.head()
```

```python
# read csv
school_table = pd.read_csv("/Users/JA/Desktop/ClassRepo/NU-VIRT-DATA-PT-02-2023-U-LOLC/02-Homework/04-Pandas/Starter_Code/Resources/schools_complete.csv")
students_table = pd.read_csv("/Users/JA/Desktop/ClassRepo/NU-VIRT-DATA-PT-02-2023-U-LOLC/02-Homework/04-Pandas/Starter_Code/Resources/students_complete.csv")

school_data_complete = pd.merge(students_table, school_table, how="left", on=["school_name", "school_name"])
df = school_data_complete




# create a variable to hold the dataframe grouped by 'school name'
df_grouped_by_school = df.groupby(["school_name"])

# create a variable that holds district type and shows  school_type
school_type = school_table.set_index(["school_name"])["type"]



# use 'size' function to hold the size of schools in total_students_count variable
total_students_count = df_grouped_by_school["size"].size().head(15)

# use 'unique' function to hold the total budget for each school
total_school_budget = df_grouped_by_school["budget"].unique().head(15)

# create variable per_student_budget to hold the average budget for each student at a school
per_student_budget = total_school_budget/total_students_count

# create a variable average_math_score that holds the average reading score for each school
# use .mean function over df_grouped_by_name dataframe
average_math_score = round(df_grouped_by_school["math_score"].mean().head(15), 2)

# create a variable average_math_score that holds the average reading score for each school
# use .mean function over df_grouped_by_name dataframe
average_reading_score = round(df_grouped_by_school["reading_score"].mean().head(15), 2)

# create a variable that hold the number that passed_math_pct for each school
# use a conditional to display values greater than 70 and groupby school_name
# use count function to find that count and divide by total_students_count for each school

passed_math_pct = df[df['math_score'] >= 70].groupby(["school_name"])['student_name'].count()/total_students_count * 100


# create a variable that hold the number that passed_reading_pct for each school
# use a conditional to display values greater than 70 and groupby school_name
# use count function to find that count and divide by total_students_count for each school
passed_reading_pct = df[df['reading_score'] >= 70].groupby(["school_name"])['student_name'].count()/total_students_count * 100

# create a variable that hold the number that passed_both_pct for each school
# use a conditional to display values greater than 70 and groupby school_name
# use count function to find that count and divide by total_students_count for each school
passed_both_pct = df[(df['reading_score'] >= 70) & (df['math_score'] >= 70)].groupby(["school_name"])['student_name'].count()/total_students_count * 100
```

```python
school_summary_df = pd.DataFrame(
                            {"School Type" : school_type,
                             "Total_Students" : total_students_count,
                             "Total School Budget" : total_school_budget,
                             "Per_Student_Budget" : per_student_budget,
                             "Average Math Score" : average_math_score,
                             "Average Reading Score" : average_reading_score,
                             "% Passing Math" : passed_math_pct,
                             "% Passing Reading" : passed_reading_pct,
                             "% Overall Passing" : passed_both_pct
                            })

# Display the DataFrame

school_summary_df.index.name = None
school_summary_df.head(15)
```

```python
# use a conditional (== to 'grade') to filter based on grade
ninth_graders_reading = df[(df["grade"] == "9th")]
tenth_graders_reading = df[(df["grade"] == "10th")]
eleventh_graders_reading = df[(df["grade"] == "11th")]
twelfth_graders_reading = df[(df["grade"] == "12th")]

# Group by "school_name" and take the mean of each.
ninth_graders_scores_reading = ninth_graders.groupby(['school_name'])['reading_score'].mean()
tenth_graders_scores_reading = tenth_graders.groupby(['school_name'])['reading_score'].mean()
eleventh_graders_scores_reading = eleventh_graders.groupby(['school_name'])['reading_score'].mean()
twelfth_graders_scores_reading = twelfth_graders.groupby(['school_name'])['reading_score'].mean()


# Combine each of the scores above into single DataFrame called `reading_scores_by_grade`
reading_scores_by_grade_df = pd.DataFrame(
                            {"9th" : ninth_graders_scores_reading,
                             "10th" : tenth_graders_scores_reading,
                             "11th" : eleventh_graders_scores_reading,
                             "12th" : twelfth_graders_scores_reading
                            })

# Data wrangling and display
reading_scores_by_grade_df.index.name = None
reading_scores_by_grade_df.head(15)
```

```python
# ---------------------------------------------------------------------------------
# SCORES BY SCHOOL SPRENDING
# ---------------------------------------------------------------------------------
```

```python
# Establish the bins 
spending_bins = [0, 585, 630, 645, 680]
labels = ["<$585", "$585-630", "$630-645", "$645-680"]

# Create a copy of the school summary since it has the "Per Student Budget" 
school_spending_df = school_summary_df.copy()

# Use `pd.cut` to categorize spending based on the bins.
school_spending_df["Spending Ranges (Per Student)"] = pd.cut(school_spending_df.Per_Student_Budget, spending_bins, labels = labels)
school_spending_df
```

```python
#  Calculate averages for the desired columns. 
spending_math_scores = school_spending_df.groupby(["Spending Ranges (Per Student)"]).mean()["Average Math Score"]
spending_reading_scores = school_spending_df.groupby(["Spending Ranges (Per Student)"]).mean()["Average Reading Score"]
spending_passing_math = school_spending_df.groupby(["Spending Ranges (Per Student)"]).mean()["% Passing Math"]
spending_passing_reading = school_spending_df.groupby(["Spending Ranges (Per Student)"]).mean()["% Passing Reading"]
overall_passing_spending = school_spending_df.groupby(["Spending Ranges (Per Student)"]).mean()["% Overall Passing"]

# Assemble into DataFrame
spending_summary_df = pd.DataFrame(
                            {"Average Math Score" : spending_math_scores,
                             "Average Reading Score" : spending_reading_scores,
                             "% Passing Math" : spending_passing_math,
                             "% Passing Reading" : spending_passing_reading,
                             "% Overall Passing" : overall_passing_spending
                            }) 

# Display results
spending_summary_df
```

```python
# ---------------------------------------------------------------------------------
# SCORES BY SCHOOL SIZE
# ---------------------------------------------------------------------------------
```

```python
# Establish the bins.
school_size_bins = [0, 1000, 2000, 5000]
school_size_labels = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]

# Categorize the spending based on the bins
# Use `pd.cut` on the "Total Students" column of the `per_school_summary` DataFrame.

school_summary_df["School Size"] = pd.cut(school_spending_df.Total_Students, school_size_bins, labels = school_size_labels)
school_summary_df["School Size"]

# Calculate averages for the desired columns. 
size_math_scores = school_summary_df.groupby(["School Size"]).mean()["Average Math Score"]
size_reading_scores = school_summary_df.groupby(["School Size"]).mean()["Average Reading Score"]
size_passing_math = school_summary_df.groupby(["School Size"]).mean()["% Passing Math"]
size_passing_reading = school_summary_df.groupby(["School Size"]).mean()["% Passing Reading"]
size_overall_passing = school_summary_df.groupby(["School Size"]).mean()["% Overall Passing"]

# Create a DataFrame called `size_summary` that breaks down school performance based on school size (small, medium, or large).
# Use the scores above to create a new DataFrame called `size_summary`
size_summary_df = pd.DataFrame(
                            {"Average Math Score" : size_math_scores,
                             "Average Reading Score" : size_reading_scores,
                             "% Passing Math" : size_passing_math,
                             "% Passing Reading" : size_passing_reading,
                             "% Overall Passing" : size_overall_passing
                            }) 

# Display results
size_summary_df
```
